```python
#atm menu
Cur_bal = 10000.0
print("Welcome to the ATM")
print( "ATM MENU" )
print( "1: Check Balance ")
print( "2: Deposit Money")
print( "3: Withdraw Money")
print( "4: exit")
while True:
    choice = float(input("Enter the choice as per the requirements between 1-4"))
    if choice == 1:
        print("Current Balance is =" , Cur_bal)
        print("Thanks for using the ATM")
        break
    elif choice == 2:
        dep_mon = float(input("Enter money to be deposited"))
        if dep_mon > 0:
           Cur_bal += dep_mon
           print("Money Deposited")
           print("Current Balance after Deopsiting Money is -" , Cur_bal)
           print("Thanks for using the ATM")
           break
    elif choice == 3:
        Wid_mon = float(input("Enter money to be withdrawn"))
        if Wid_mon <= Cur_bal:
            Cur_bal -= Wid_mon
            print("the money is withdrawn")
            print("Balance after withdrawing money -" , Cur_bal)
            print("Thanks for using the ATM")
            break
        else:
            print("Insufficient Balance in Bank Account ")
            print("Enter your Choice and Money again")
    elif choice == 4:
        print ("exit")
        print ("Thanks for using the ATM")        
        break
    else:
        if choice > 4:
            print("Invalid Choice, Enter the choice again.")
        
 
```
